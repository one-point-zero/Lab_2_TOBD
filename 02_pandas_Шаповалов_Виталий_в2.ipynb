### Объединение таблиц `pd.DataFrame`

5.1 При помощи объединения таблиц, создайте `DataFrame`, состоящий из четырех столбцов: `id`, `name`, `user_id`, `rating`. Рецепты, на которые не оставлен ни один отзыв, должны отсутствовать в полученной таблице. Подтвердите правильность работы вашего кода, выбрав рецепт, не имеющий отзывов, и попытавшись найти строку, соответствующую этому рецепту, в полученном `DataFrame`.

```python
new_recipes = recipes[['id', 'name']] 
new_reviews = reviews[['user_id', 'recipe_id','rating']]
new_frame = pd.merge(new_recipes, new_reviews, left_on = 'id', right_on = 'recipe_id')
new_frame.drop('recipe_id', axis = 1, inplace = True)
print(new_frame)
not_out = ~recipes['id'].isin(reviews['recipe_id'])
print(np.random.choice(new_frame['id'][not_out])) # проверяем правильность работы кода (выдаёт ошибку, значит, все работает правильно)
```

5.3. Выясните, рецепты, добавленные в каком году, имеют наименьший средний рейтинг?

```python
date_rec = pd.merge(recipes[['id', 'submitted']], reviews[['recipe_id','rating']], left_on = 'id', right_on = 'recipe_id')
date_rec = date_rec.drop_duplicates()
del date_rec[date_rec.columns[0]]
del date_rec[date_rec.columns[1]]
date_rec['submitted'] = date_rec['submitted'].dt.to_period("Y")
mid_rating = date_rec.groupby('submitted')['rating'].mean()
min_r = mid_rating[mid_rating[mid_rating == min(mid_rating)].index] 
min_r # год рецептов с наименьшим рейтингом
```
